### Q1 . Perform the following operations on the given dataset:

```python
# Generating the data set
import pandas as pd

data = {
    'Name':['Mudit Chauhan','Seema Chopra','Rani Gupta','Aditya Narayan','Sanjeev Sahni','Prakash Kumar','Ritu Agarwal','Akshay Goel','Meeta Kulkarni','Preeti Ahuja','Sunil Das Gupta','Sonali Sapre','Rashmi Talwar'],
    'Birth_Month':['December','January','March','October','February','December','September','August','July','November','April','January','June'],
    'Gender':['M','F','F','M','M','M','F','M','F','F','M','F','F'],
    'Pass_Division':['III','II','I','I','II','III','I','I','II','II','III','I','III']
}


df = pd.DataFrame(data)
 
df.to_excel('dataset.xlsx',index=False)
```

a) Data transformation, aggregation, slicing, and indexing.

```python
import pandas as pd

# Read the dataset from the generated Excel file
df = pd.read_excel('dataset.xlsx')

# Example of data transformation: Extract first names from the 'Name' column
df['First_Name'] = df['Name'].apply(lambda x: x.split()[0])

# Example of data aggregation: Group by 'Birth_Month' and count the number of entries
birth_month_counts = df['Birth_Month'].value_counts()

# Example of data slicing and indexing: Select rows with 'Gender' equals 'M'
male_data = df[df['Gender'] == 'M']

# Display the results
print("Extracted First Names:")
print(df[['Name', 'First_Name']])

print("\nBirth Month Counts:")
print(birth_month_counts)

print("\nMale Participants:")
print(male_data)
```

b) Prepare data for visualization.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Prepare data for a bar chart showing the frequency of each class label in 'Pass_Division'
pass_division_counts = df['Pass_Division'].value_counts()
```

c) Plot bar chart to show the frequency of each class label in the data.

```python
plt.figure(figsize=(8, 5))
pass_division_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency of Pass Division')
plt.xlabel('Pass Division')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.savefig('Frequency of Pass Division')
plt.show()
```

d) Draw a Pie chart for Pass_division.

```python
plt.figure(figsize=(8, 5))
pass_division_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Pass Division Distribution')
plt.ylabel('')
plt.savefig('Pie Chart for Pass_division')
plt.show()
```

e) Plot density distribution for Birth_Month.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dictionary to map month names to numerical values
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}

# Convert 'Birth_Month' to numerical values (Since kdeplot needs numeric or datetime data but birthmonth is string)
df['Birth_Month_Num'] = df['Birth_Month'].map(month_mapping)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.kdeplot(df['Birth_Month_Num'], fill=True, color="skyblue")
plt.title('Density Distribution of Birth Month')
plt.xlabel('Birth Month (Numerical)')
plt.ylabel('Density')
plt.xticks(range(1, 13), rotation=45)
plt.savefig('Density Distribution of Birth Month')
plt.show()
```
